```python
from keplergl import KeplerGl
import pandas as pd
```

```python
df = pd.read_csv('/Users/carboni/Downloads/combined_buste.csv')
```

```python
df.head()
```

```python
def date_split(date:str)-> (str,str,str):
    if type(date) is not str:
        return float('nan'),float('nan'),float('nan')
    splits = date.split('-')
    vals = splits + [float('nan')] * (3 - len(splits))
    return [float(v) for v in vals]

df['year'], df['month'], df['day'] = zip(*df['normalized_date'].apply(date_split))

df = df[df['year'] >= 1870.]
```

```python
df = df.drop(['month','day', 'year'], 1)
```

```python
df['normalized_date'] = df['normalized_date'].astype(str) + 'T00:00:00Z'
```

```python
df.head(3)
```

## Compute pivot

number of cities and number of type

```python
pivot_city = pd.pivot_table(df, values='image_url', index='City', aggfunc=pd.Series.nunique)
```

```python
pivot_city = pivot_city.rename(columns={"image_url": "City_number"})
```

```python
df = pd.merge(df, pivot_city, on=['City'])
```

```python
df.head(4)
```

```python
pivot_type = pd.pivot_table(df, values='image_url', index='Journal Type', aggfunc=pd.Series.nunique)
```

```python
pivot_type = pivot_type.rename(columns={"image_url": "Type_number"})
```

```python
df = pd.merge(df, pivot_type, on=['Journal Type'])
```

```python
df.head(4)
```
