```python
from keplergl import KeplerGl
import pandas as pd
```

```python
df = pd.read_csv('/Users/carboni/Downloads/avant.csv')
```

```python
df.head()
```

```python
def date_split(date:str)-> (str,str,str):
    if type(date) is not str:
        return float('nan'),float('nan'),float('nan')
    splits = date.split('-')
    vals = splits + [float('nan')] * (3 - len(splits))
    return [float(v) for v in vals]

df['year'], df['month'], df['day'] = zip(*df['Date'].apply(date_split))

df1 = df[df['year'] <= 1950.]
```

```python
df1 = df1.drop(['month','day', 'year'], 1)
```

```python
df1['Date'] = df1['Date'].astype(str) + 'T00:00:00Z'
```

```python
df1.head(3)
```

```python
df1 = df1.drop('notice', axis=1)
```

## Compute pivot

number of cities and number of type

```python
pivot_city = pd.pivot_table(df1, values='image_url', index='City', aggfunc=pd.Series.nunique)
```

```python
pivot_city = pivot_city.rename(columns={"image_url": "City_number"})
```

```python
df1 = pd.merge(df1, pivot_city, on=['City'])
```
