```python
#!pip install geopy
import pandas as pd
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
from tqdm import tqdm
tqdm.pandas()
```

```python
df = pd.read_csv('/Users/carboni/Downloads/FJC data.csv')
```

```python
df.head()
```

```python
df = df.dropna(subset=['https://visualcontagions.net/schema#City'])
```

## If you have both city and country information in your data

```python
geolocator = Nominatim(user_agent="geoapiExercises")

# Function for geocoding
def geocode(city, country):
    try:
        location = geolocator.geocode(f"{city}, {country}")
        return location.latitude, location.longitude
    except:
        return None, None

# Create a rate limiter to respect usage limits
geocode_with_rate_limiter = RateLimiter(geocode, min_delay_seconds=1)

# Using a dictionary to cache results
cache = {}

def cached_geocode(city, country):
    key = (city, country)
    if key not in cache:
        cache[key] = geocode_with_rate_limiter(city, country)
    return cache[key]
```

```python
df['coord'] = df.progress_apply(lambda row: cached_geocode(row['https://visualcontagions.net/schema#City'], row['Country']), axis=1)
```

## If you have only the city information in your data

```python
def geocode(city):
    try:
        location = geolocator.geocode(city)
        return location.latitude, location.longitude
    except:
        return None, None

# rate limiter 
geocode_with_rate_limiter = RateLimiter(geocode, min_delay_seconds=1)

# cache results
cache = {}

def cached_geocode(city):
    if city not in cache:
        cache[city] = geocode_with_rate_limiter(city)
    return cache[city]
```

```python
df['coord'] = df['https://visualcontagions.net/schema#City'].progress_apply(cached_geocode)
```

## Common operation

```python
df['coord'] = df['coord'].apply(lambda coord: ', '.join(map(str, coord)) if coord else None)
```

```python
df[['lat', 'long']] = df['coord'].str.split(',', n=1, expand=True)
```
