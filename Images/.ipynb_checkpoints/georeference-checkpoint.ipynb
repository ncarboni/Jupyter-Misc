```python
#!pip install geopy
import pandas as pd
from geopy.geocoders import Nominatim
from tqdm import tqdm
tqdm.pandas()
```

```python
df = pd.read_csv('/Users/carboni/Downloads/Leopard.csv')
```

```python
df.head()
```

```python
df = df.dropna(subset=['City'])
```

```python
geolocator = Nominatim(user_agent="geoapiExercises")

def geocode(city, country):
    try:
        location = geolocator.geocode(f"{city}, {country}")
        return location.latitude, location.longitude
    except:
        return None, None
```

```python
df['coords'] = df.progress_apply(lambda row: geocode(row['City'], row['Country']), axis=1)
```

```python
df.head()
```

```python
df[['latitude', 'longitude']] = pd.DataFrame(df['coords'].tolist(), index=df.index)
```

```python
df.to_csv('georeferenced.csv', index=False)
```
