```python
path = r'/Users/carboni/Documents/UNIGE/Corpus/bustes/merged/' # use your path
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0, delimiter=',')
    li.append(df)

df2 = pd.concat(li, axis=0, ignore_index=True)
```

```python
df2.head()
```

```python
df2 = df2.drop_duplicates(subset=['image_url'])
```

## Clean dates

```python
df2["normalized_date"].unique()
```

```python
df2[['normalized_date1','normalized_date2']]=df2.normalized_date.str.split(pat=',',expand=True)
```

```python
df2.drop(['normalized_date2','normalized_date'], axis=1, inplace=True)
```

```python
df2.rename(columns = {'normalized_date1':'normalized_date'}, inplace = True)
```

```python
df2.head(2)
```

```python
re_express1 = re.compile('([\d]{4}).([\d]{2})$')
re_express2 = re.compile('([\d]{4})$')
re_express3 = re.compile('([\d]{4}).([\d]{2}).([\d]{2})$')
df2['normalized_date'] = df2['normalized_date'].str.replace(re_express2, r'\1-01-01')
df2['normalized_date'] = df2['normalized_date'].str.replace(re_express1, r'\1-\2-01')
df2['normalized_date'] = df2['normalized_date'].str.replace(re_express3, r'\1-\2-\3')
```

```python
df2.to_csv('/Users/carboni/Documents/UNIGE/Corpus/bustes/merged/combined_buste', index=False)
```
