```python
import os
import pandas as pd
import warnings
import glob
```

```python
path = r'/Users/carboni/Downloads/merge_csv' # use your path
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0, delimiter=',', dtype ='str')
    li.append(df)

df2 = pd.concat(li, axis=0, ignore_index=True)
```

```python
df2.to_csv('/Users/carboni/Downloads/tate_combined.csv', index=False)
```
