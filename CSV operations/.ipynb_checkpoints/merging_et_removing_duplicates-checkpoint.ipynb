## Merging CSV

```python
import os
import pandas as pd
import warnings
import glob
```

```python
path = r'/Users/carboni/Documents/UNIGE/cars/CSV/merged' # use your path
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0, delimiter=',')
    li.append(df)

df2 = pd.concat(li, axis=0, ignore_index=True)
```

## Drop duplicates

```python
df2.drop_duplicates(subset=['image_url'])
```

```python
df2.to_csv('/Users/carboni/Documents/UNIGE/cars/CSV/merged/combined_cars.csv', index=False)
```
