# Pivot Notebook - Explore results

To be used to create pivot table for the explored corpus of VC. It opens a file and create pivot table according to
+ City
+ Journal
+ Type

## Computing Pivot

```python
import pandas as pd
```

```python
df = pd.read_csv("/Users/carboni/Downloads/cluster_164.csv")
```

```python
df.head()
```

### Quantify how many Images have been circulating on the same city

```python
pivot_city = pd.pivot_table(df, values='image_url', index='City', aggfunc=pd.Series.nunique)
```

```python
pivot_city.head(20)
```

```python
pivot_city = pivot_city.rename(columns={"image_url": "n_image_per_city"})
```

```python
df = pd.merge(df, pivot_city, on=['City'])
df.head(3)
```

## Neo4j

### DateTime

```python
df['Date'] =  pd.to_datetime(df['Date'])
```

```python
df.head(6)
```

```python
df.info()
```

#### Only year

```python
df['year'] = pd.DatetimeIndex(df['Date']).year.astype(str)
```

```python
df['year'] = df['year'].str.replace('.0','')
```

```python
df.head(3)
```

### WKT to lat/long 

```python
## not possible as not everything has coordinates
# from shapely import wkt
# import geopandas as gpd
# df['wkt'] = df['wkt'].apply(wkt.loads)
# gdf = gpd.GeoDataFrame(df, geometry='wkt')
# df['lat'] = gdf.geometry.x
# df['long'] = gdf.geometry.y
```

```python
df[['lat', 'long']] = df['wkt'].str.split(' ', n=2, expand=True)
```

```python
df['lat'] = df['lat'].str.replace('POINT','')
df['lat'] = df['lat'].str.replace('(','')
df['long'] = df['long'].str.replace(')','')
```

```python
df.head(2)
```

### lowercase columns

```python
# rename journal type and eliminate space
df.rename(columns = {'Journal Type':'Journal_Type'}, inplace = True)
```

```python
df.columns=map(str.lower, df.columns)
```

```python
df.head(3)
```

```python
df.info()
```

```python
df = df.drop(columns=['canvas_number', 'notice'])
```
