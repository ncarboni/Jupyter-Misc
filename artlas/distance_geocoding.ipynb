```python
import pandas as pd
import geopy.distance
```

```python
df = pd.read_csv("/Users/carboni/Downloads/distances%2Dkm2%2Dcsv%2Ecsv.txt")
```

```python
df.head()
```

```python
df.drop(['reverse_coordinates'], axis=1, inplace=True)
```

```python
df.dropna(subset = ["x"], inplace=True)
```

```python
df = df.dropna(subset=['new_long','city_x'])
```

```python
def calc_distance(x):
    site_coords = (x.city_y, x.city_x)
    place2_coords = (x.new_lat, x.new_long)
    return geopy.distance.distance(site_coords, place2_coords).km

df['distance_km2'] = df.apply(calc_distance, axis = 1)
```
