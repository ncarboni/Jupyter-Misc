```python
def download_and_rename_image(image_url, folder_path):
    try:
        response = requests.get(image_url, stream=True)
        
        # Check if the response status code is 200 (OK)
        if response.status_code == 200:
            image_name = re.sub(r'[^a-zA-Z0-9]', '_', image_url)  # Remove non-alphanumeric characters from URL
            image_name = f"{image_name}.jpg"  # Add .jpg extension
            image_path = os.path.join(folder_path, image_name)

            # Check if the image is already downloaded to avoid redundant downloads
            if not os.path.exists(image_path):
                with open(image_path, 'wb') as f:
                    for chunk in response.iter_content(chunk_size=8192):
                        f.write(chunk)
                print(f"Downloaded: {image_url}")
                return image_path  # Return the image path
            else:
                print(f"Already downloaded: {image_url}")
                return image_path  # Return the image path if already downloaded
        else:
            print(f"Error: {image_url} returned status code {response.status_code}")

    except Exception as e:
        print(f"Error downloading {image_url}: {e}")

    return None  # Return None if download fails
```

```python
folder_path = 'image_download'
```

```python
# Create a list to store the image paths
image_paths = []

for image_url in image_urls:
    downloaded_path = download_and_rename_image(image_url, folder_path)
    if downloaded_path:
        image_paths.append(downloaded_path)  # Append the image path
```

```python
# Add a new column 'image_path' to the original DataFrame 'df'
df['image_path'] = image_paths
```

## Download 100 then pause

```python
def download_and_rename_pdfs(pdf_urls, download_folder):
    if not os.path.exists(download_folder):
        os.makedirs(download_folder)
    
    for i, pdf_url in enumerate(pdf_urls, 1):
        try:
            response = requests.get(pdf_url)
            response.raise_for_status()  # Check for successful response
            
            # Get the file name from the URL
            url_path = urlparse(pdf_url).path
            filename = os.path.basename(url_path)
            
            # Construct the full path to save the PDF
            pdf_path = os.path.join(download_folder, filename)
            
            # Check if the PDF file already exists in the download folder
            if not os.path.exists(pdf_path):
                # Download the PDF and save it with the new name
                with open(pdf_path, "wb") as pdf_file:
                    pdf_file.write(response.content)
                
                print(f"Downloaded and saved: {pdf_path}")
            else:
                print(f"Skipped download (already exists): {pdf_path}")
            
            # Pause for 30 seconds after every 100 downloads
            if i % 100 == 0:
                print(f"Pausing for 30 seconds...")
                time.sleep(30)
        
        except Exception as e:
            print(f"Error downloading {pdf_url}: {e}")

# Call the function to start downloading and renaming
download_folder = "fedora"
download_and_rename_pdfs(pdf_urls, download_folder)
```
