```python
import urllib.request
import uuid
import time
import pandas as pd
import os
import requests
import re
```

# Download Images

```python
df = pd.read_csv('/Users/carboni/Downloads/Image Series Bosboom Church Interior.csv')
```

```python
image_urls = df['image_url'].tolist() 
```

```python
df.head()
```

```python
def download_and_rename_image(image_url, folder_path):
    try:
        response = requests.get(image_url, stream=True)
        if response.status_code == 200:
            image_name = re.sub(r'[^a-zA-Z0-9]', '_', image_url)  # Remove non-alphanumeric characters from URL
            image_name = f"{image_name}.jpg"  # Add .jpg extension
            image_path = os.path.join(folder_path, image_name)

            # Check if the image is already downloaded to avoid redundant downloads
            if not os.path.exists(image_path):
                with open(image_path, 'wb') as f:
                    for chunk in response.iter_content(chunk_size=8192):
                        f.write(chunk)
                print(f"Downloaded: {image_url}")
                return True

    except Exception as e:
        print(f"Error downloading {image_url}: {e}")

    return False

def download_images(image_urls, folder_path, batch_size=100, pause_duration=10):
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)

    image_count = 0

    for image_url in image_urls:
        downloaded = download_and_rename_image(image_url, folder_path)
        if downloaded:
            image_count += 1

        # Pause every 'batch_size' images downloaded
        if image_count % batch_size == 0:
            print(f"Pausing for {pause_duration} seconds...")
            time.sleep(pause_duration)
```

```python
download_images(image_urls, 'image_download', batch_size=100, pause_duration=10)
```

## Download PDFs

```python
df = pd.read_csv('/Users/carboni/Downloads/test_no_drive.csv')
```

```python
df.head()
```

```python
pdf_urls = df['Media URL'].tolist() 
```

```python
from urllib.parse import urlparse

def download_and_rename_pdfs(pdf_urls, download_folder):
    if not os.path.exists(download_folder):
        os.makedirs(download_folder)
    
    for pdf_url in pdf_urls:
        try:
            response = requests.get(pdf_url)
            response.raise_for_status()  # Check for successful response
            
            # Get the file name from the URL
            url_path = urlparse(pdf_url).path
            filename = os.path.basename(url_path)
            
            # Construct the full path to save the PDF
            pdf_path = os.path.join(download_folder, filename)
            
            # Download the PDF and save it with the new name
            with open(pdf_path, "wb") as pdf_file:
                pdf_file.write(response.content)
                
            print(f"Downloaded and saved: {pdf_path}")
        
        except Exception as e:
            print(f"Error downloading {pdf_url}: {e}")


# Call the function to start downloading and renaming

download_folder = "pdf_download"
download_and_rename_pdfs(pdf_urls, download_folder)
```

## download only if it does not exist

```python
import pandas as pd
import os
import requests
from urllib.parse import urlparse

df = pd.read_csv('/Users/carboni/Downloads/test_chile.csv')

pdf_urls = df['Media URL'].tolist() 

def download_and_rename_pdfs(pdf_urls, download_folder):
    if not os.path.exists(download_folder):
        os.makedirs(download_folder)
    
    for pdf_url in pdf_urls:
        try:
            response = requests.get(pdf_url)
            response.raise_for_status()  # Check for successful response
            
            # Get the file name from the URL
            url_path = urlparse(pdf_url).path
            filename = os.path.basename(url_path)
            
            # Construct the full path to save the PDF
            pdf_path = os.path.join(download_folder, filename)
            
            # Check if the PDF file already exists in the download folder
            if not os.path.exists(pdf_path):
                # Download the PDF and save it with the new name
                with open(pdf_path, "wb") as pdf_file:
                    pdf_file.write(response.content)
                
                print(f"Downloaded and saved: {pdf_path}")
            else:
                print(f"Skipped download (already exists): {pdf_path}")
        
        except Exception as e:
            print(f"Error downloading {pdf_url}: {e}")

# Call the function to start downloading and renaming
download_folder = "pdf_download"
download_and_rename_pdfs(pdf_urls, download_folder)
```
