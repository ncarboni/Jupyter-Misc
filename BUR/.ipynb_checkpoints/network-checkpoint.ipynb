```python
def plot_g_pyviz(M, name='out.html', height='300px', width='500px'):
    g = M.copy() #
    net = pvnet.Network(notebook=True, directed=True, height=height, width=width)
    opts = '''
        var options = {
          "physics": {
            "forceAtlas2Based": {
              "gravitationalConstant": -100,
              "centralGravity": 0.11,
              "springLength": 100,
              "springConstant": 0.09,
              "avoidOverlap": 1
            },
            "minVelocity": 0.75,
            "solver": "forceAtlas2Based",
            "timestep": 0.22
          }
        }
    '''

    net.set_options(opts)
    net.from_nx(g)
    return net.show(name)

plot_g_pyviz(M)
```

```python
# load custom version of pyvis
from dimcli.utils.networkviz import NetworkViz


def build_visualization(table):
    """
    Return a network visualization object from a collaborators dataframe
    The object can be then displayed/saved with `g.show(f"network.html")`
    """

    # set up dataviz
    g = NetworkViz(notebook=True, width="100%", height="800px")
    g.toggle_hide_edges_on_drag(False)
    g.barnes_hut()
    g.repulsion(300)
    #g.heading = f"Collaboration network for '{GRIDID} - {gridname}'"
    # reuse plotly color palette
    palette = px.colors.diverging.Temps
    # g.show_buttons() # in html-standalone mode, this command shows viz controls


    #
    # create nodes and edges
    #

    # remove duplicates from nodes
    nodes = table.drop_duplicates(subset ="envelope_receiver_person_korbo_label", keep = 'first')
    # remove internal collaborations stats
    edges = table[(table['envelope_receiver_person_korbo_label'] != table['envelope_receiver_person_korbo_label'])]


    #
    # add nodes
    #

    for index, row in nodes.iterrows():

        # calc size based on level
        maxsize = int(nodes['level'].max()) + 1
        if row['envelope_receiver_person_korbo_label'] == GRIDID:
            size = maxsize
        else:
            size = maxsize - row['level']

        # calc color based on level
        if row['envelope_receiver_person_korbo_label'] == GRIDID:
            color = palette[0]
        else:
            color = palette[row['level'] * 2]

        g.add_node(
            n_id = row['envelope_receiver_person_korbo_label'],
            label = row['name'],
            title = f"<h4>{row['name']}<br>{row['city_name']}, {row['country_name']}<br> - {row['envelope_receiver_person_korbo_label']}</h4>",
            value = size,
            color = color,
            borderWidthSelected = 5,
            shape = "dot",
        )


    #
    # add edges
    #

    edges_maxcount = edges['count'].max()

    for index, row in edges.iterrows():
      g.add_edge(row['envelope_receiver_person_korbo_label'], row['envelope_receiver_person_korbo_label'],
                 value = float(row['count']) / edges_maxcount,
                 label=int(row['count']),
                 arrows="none"
                )

    # add tooltips with adjancent links info
    neighbor_map = g.get_adj_list()
    for node in g.nodes:
        neigh = neighbor_map[node["id"]]
        labels = [nodes[nodes['envelope_receiver_person_korbo_label'] == x].iloc[0]['name'] for x in neigh]
        node["title"] += "Links:<li>" + "</li><li>".join(labels)

    return g

#
# finall, run the viz builder
#
#collaborators = recursive_network(GRIDID, maxlevel=2)
#g = build_visualization(collaborator)
g.show(f"network_{GRIDID}.html")
```
