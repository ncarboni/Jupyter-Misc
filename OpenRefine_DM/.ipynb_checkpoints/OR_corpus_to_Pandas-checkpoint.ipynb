```python
import pandas as pd
import glob
```

```python
path = r'/Users/nicola/Downloads/beatrice' # use your path
all_files = glob.glob(path + "/*.txt")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0, delimiter='\t', low_memory=False, dtype=object)
    li.append(df)

frame = pd.concat(li, axis=0, ignore_index=True)
```

```python
frame.to_csv('combined.csv')
```
