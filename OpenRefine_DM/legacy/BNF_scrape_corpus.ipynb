```python
import re
from bs4 import BeautifulSoup
import requests
from functools import reduce
```

```python
url = "https://visualcontagions.unige.ch/explore/corpus/dataset/150/?page=12"
req = requests.get(url)
soup = BeautifulSoup(req.content, 'html.parser')
```

```python
with open("output12.txt", "w", encoding='utf-8') as file:
    for el in soup.find_all(class_="manifest-link"):
            file.write(el.get_text())
```
