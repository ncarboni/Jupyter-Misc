```python
import movingpandas as mpd
import geopandas as gpd
import pandas as pd
from shapely.wkt import loads
```

```python
df = pd.read_csv('/Users/carboni/Downloads/Avantguarde.csv')
```

```python
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
transformed_data = []

for cluster, group_df in df.groupby('numero_cluster'):
    group_df = group_df.sort_values(by='Date')
    group_df['origin'] = group_df['wkt']
    group_df['destination'] = group_df['wkt'].shift(-1).fillna(group_df['wkt'].iloc[-1])
    transformed_data.append(group_df)

# Concatenate all transformed groups
transformed_df = pd.concat(transformed_data)
```

```python
transformed_df = transformed_df.sort_values(by='numero_cluster')
```

```python
transformed_df.head()
```

```python
transformed_df["origin"] = transformed_df["origin"].apply(loads)
transformed_df["destination"] = transformed_df["destination"].apply(loads)
transformed_df["Date"] = pd.to_datetime(transformed_df["Date"])
```

```python
gdf = gpd.GeoDataFrame(transformed_df, geometry='origin')
gdf.set_crs("EPSG:4326", inplace=True)  # Set the CRS
```

```python
gdf = gdf.dropna(subset=['origin'])
```

```python
trajectories = []
for name, group in gdf.groupby("numero_cluster"):
    group = group.set_index("Date").sort_index()  # Set 'Date' as index and sort by it
    trajectory = mpd.Trajectory(group, traj_id=name)
    trajectories.append(trajectory)

plot = mpd.TrajectoryCollection(trajectories).hvplot(title="Trajectories")
```

```python
import hvplot
plot = mpd.TrajectoryCollection(trajectories).hvplot(title="Trajectories")
hvplot.show(plot)
```
