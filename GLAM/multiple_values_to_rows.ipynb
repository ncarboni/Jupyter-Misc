```python
import pandas as pd
```

```python
df = pd.read_csv('/Users/carboni/Downloads/MAH Artworks.csv')
```

```python
df.head()
```

```python
new_rows = []

for _, row in df.iterrows():
    ids_museum = str(row['ID artiste MuseumPlusMAH']).split(' $$ ')
    auteur_intervenant_list = str(row['Auteur/Intervenant']).split(' $$ ')
    
    for id_museum in ids_museum:
        for auteur_intervenant in auteur_intervenant_list:
            if f'ID: {id_museum}' in auteur_intervenant:
                new_row = row.copy()
                new_row['ID artiste MuseumPlusMAH'] = id_museum
                new_row['Auteur/Intervenant'] = auteur_intervenant
                new_rows.append(new_row)

result_df = pd.DataFrame(new_rows)
```

```python
result_df = result_df.sort_values(by='ID œuvre')
```

```python
result_df.to_csv('artwork_full.csv', index=False)
```
